```python
import pandas as pd
from sentence_transformers import InputExample,util
```

```python
df = pd.read_csv('annotated_data.csv')
```

```python
df
```

```python
df = df.dropna()
df = df.drop_duplicates(['history_opinion'])
df = df.reset_index()
```

```python
df
```

```python
# df['label'] = df['source_citation'].apply(lambda x : 1)
```

```python
def negative_sampling(df):
  n_rows = df.shape[0]
  neg = []
  for i in range(n_rows):
    neg.append(df.iloc[(i + 1)% n_rows]['history_opinion'])
  df['negative_opinion'] = neg
      # print(new_row)
  return df
```

```python
df_new = negative_sampling(df)
```

```python
df_new
```

```python

n_train = int(df.shape[0]* 0.8)
n_val = int(df.shape[0] * 0.1)
n_test = n_val
train_data = df[:n_train]
val_data = df[n_train:n_train + n_val]
test_data = df[n_train + n_val:]
# For agility we only 1/2 of our available data
```

```python
train_examples = []
for i in range(n_train):
  example = train_data.iloc[i]
  train_examples.append(InputExample(texts=[example['source_opinion'], example['history_opinion'], example['negative_opinion']]))
```

```python
val_examples = []
for i in range(n_val):
  example = val_data.iloc[i]
  val_examples.append(InputExample(texts=[example['source_opinion'], example['history_opinion'], example['negative_opinion']]))
```

```python
test_examples = []
for i in range(n_test):
  example = test_data.iloc[i]
  test_examples.append(InputExample(texts=[example['source_opinion'], example['history_opinion'], example['negative_opinion']]))
```

```python
from torch.utils.data import DataLoader

train_dataloader = DataLoader(train_examples, shuffle=True, batch_size=8)
```

```python
from sentence_transformers import losses
from sentence_transformers import SentenceTransformer, models
import torch
device = torch.device("cuda" if torch.cuda.is_available() else "cpu")

## Step 1: use an existing language model
# word_embedding_model = models.Transformer('distilroberta-base').to(device)

# ## Step 2: use a pool function over the token embeddings
# pooling_model = models.Pooling(word_embedding_model.get_word_embedding_dimension()).to(device)

# ## Join steps 1 and 2 using the modules argument
# model = SentenceTransformer(modules=[word_embedding_model, pooling_model]).to(device)

model_id = "sentence-transformers/all-MiniLM-L6-v2"
model = SentenceTransformer(model_id)

train_loss = losses.TripletLoss(model=model).to(device)

num_epochs = 10

warmup_steps = int(len(train_dataloader) * num_epochs * 0.1) #10% of train data
```

```python
model.fit(train_objectives=[(train_dataloader, train_loss)],
          epochs=num_epochs,
          warmup_steps=warmup_steps)
```

```python
torch.save(model.state_dict(), 'legal_finetuned_sbert.')
```

```python
for i in range(len(val_data)):
  source_embedding = model.encode(val_data.iloc[i]['source_opinion'])
  history_embedding = model.encode(val_data.iloc[i]['history_opinion'])
  neg_embedding = model.encode(val_data.iloc[i]['negative_opinion'])
  cosine_scores_pos = util.cos_sim(source_embedding, history_embedding)
  cosine_scores_neg = util.cos_sim(source_embedding, neg_embedding)
  print(val_data.iloc[i]['source_opinion'])
  print(val_data.iloc[i]['history_opinion'])
  print(i+1, cosine_scores_pos, cosine_scores_neg, cosine_scores_pos> cosine_scores_neg)
```
