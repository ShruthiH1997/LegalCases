# Annotated California Cases

```python
annotated_df = pd.read_csv('data/full_expanded.csv', usecols=['citation', 'case', 'shepID',
                                                              'itemID', 'date', 'court', 'parallelCitations',
                                                              'priorParallelCitations', 'relationToCurrent'])
annotated_df['citation'] = annotated_df['citation'].str.strip()
annotated_df['relationToCurrent'] = annotated_df['relationToCurrent'].apply(lambda x: ast.literal_eval(x))
annotated_df['parallelCitations'] = annotated_df['parallelCitations'].apply(lambda x: ast.literal_eval(x))
annotated_df['priorParallelCitations'] = annotated_df['priorParallelCitations'].apply(lambda x: ast.literal_eval(x))
annotated_df.info()
```

```python
annotated_df.head()
```

```python
# Strip prior citations of surrounding whitespace
for _, row in annotated_df.iterrows():
    for citation_list in row['priorParallelCitations']:
        for i, citation in enumerate(citation_list):
            citation_list[i] = citation.strip()
```

```python
# Create a set of all distinct citations in the annotated dataset
all_annotated_citations = set(annotated_df['citation'].values)
len(all_annotated_citations)
```

```python
# Create a set of all distinct citations in the unannotated dataset
all_citations = set()
for citations in df.citations:
    for citation in citations:
        all_citations.add(citation['cite'].strip())
len(all_citations)
```

```python
# Unannotated data has no lexis citations
for citation in all_citations:
    if 'LEXIS' in citation:
        print(citation)
```

```python
# How many annotated citations don't exist in our unannotated data?
citations_only_in_annotated_dataset = all_annotated_citations.difference(all_citations)
len(citations_only_in_annotated_dataset)
```

```python
# Find those cases from the annotated dataset whose citations we don't have in the unannotated set
problems_annotated_df = annotated_df[annotated_df.citation.isin(citations_only_in_annotated_dataset)]
problems_annotated_df.to_csv('unmatched_cases.csv', columns=['citation', 'case', 'shepID', 'itemID', 'date', 'court'], index=False)
problems_annotated_df
```

```python
# How many of the annotated citations exist in our unannotated data?
citations_in_both_datasets = all_citations.intersection(all_annotated_citations)
len(citations_in_both_datasets)
```

```python
# Select only those cases from the annotated dataset whose citations we have in the unannotated set
annotated_df = annotated_df[annotated_df.citation.isin(citations_in_both_datasets)]
annotated_df
```

```python
# Even in these cases, the history cases are almost always referred to by their LEXIS ids
annotated_df['priorParallelCitations'].iloc[0]
```

```python
# Filter history to contain only prior cases
def filter_non_prior(relationToCurrent, priorParallelCitations):
    filteredpriorParallelCitations = []
    for i in range(min(len(relationToCurrent), len(priorParallelCitations))):
        if relationToCurrent[i] == 'prior':
            filteredpriorParallelCitations.append(priorParallelCitations[i])
    return ['prior'] * len(filteredpriorParallelCitations), filteredpriorParallelCitations

for i, row in annotated_df.iterrows():
    relations, citations = filter_non_prior(annotated_df.loc[i, 'relationToCurrent'],
                                            annotated_df.loc[i, 'priorParallelCitations'])
    annotated_df.loc[i, 'relationToCurrent'] = relations
    annotated_df.loc[i, 'priorParallelCitations'] = citations
```

```python
# Filter out cases with mismatched history and relations
def get_case_indices_with_mismatched_relation_list(df):
    case_indices_with_mismatched_relation_list = []
    for i, row in df.iterrows():
        if len(row['relationToCurrent']) != len(row['priorParallelCitations']):
            case_indices_with_mismatched_relation_list.append(i)
    print(len(case_indices_with_mismatched_relation_list))
    return case_indices_with_mismatched_relation_list

case_indices_with_mismatched_relation_list = get_case_indices_with_mismatched_relation_list(annotated_df)
annotated_df = annotated_df.drop(case_indices_with_mismatched_relation_list)
annotated_df
```

```python
# Create a set of all distinct citations in the histories of the cases
history_citations = set()
for case_history in annotated_df['priorParallelCitations'].values:
    for case in case_history:
        try:
            for citation in case:
                history_citations.add(citation)
        except:
            pass
len(history_citations)
```

```python
# Proportion of LEXIS/Non-LEXIS citations
non_lexis_history_citations = set()
lexis_history_citations = set()
for citation in history_citations:
    if 'LEXIS' not in citation:
        non_lexis_history_citations.add(citation)
    else:
        lexis_history_citations.add(citation)
print(f'LEXIS/Non-LEXIS citations in history: {len(lexis_history_citations)}/{len(non_lexis_history_citations)}')
```

```python
# How many of these history citations exist in our unannotated data?
history_citations_in_both_datasets = all_citations.intersection(history_citations)
len(history_citations_in_both_datasets)
```

```python
# How many citations (history or current) from the annotated dataset exist in our unannotated data?
all_matching_citations = history_citations_in_both_datasets.union(citations_in_both_datasets)
len(all_matching_citations)
```

```python
# Explode the unannotated df so that each row has a unique citation 
exploded_df = df.explode('citations')
exploded_df['citations'] = exploded_df['citations'].apply(lambda x: x['cite'].strip())
exploded_df = exploded_df[exploded_df['citations'].isin(all_matching_citations)]
exploded_df.shape
```

```python
# Select only the columns we need
exploded_df_small = exploded_df[['name', 'citations', 'decision_date', 'court', 'casebody']]
exploded_df_small
```

```python
# Create a new list of annotated cases, containing only matching cases in their histories
data_all = []
for _, row in annotated_df.iterrows():
    data = {}
    data['citation'] = row['citation']
    data['priors'] = set()
    for case in row['priorParallelCitations']:
        for citation in case:
            if citation in all_matching_citations:
                data['priors'].add(citation)
    if len(data['priors']) > 0:
        data_all.append(data)
len(data_all)
```

```python
# Notice how sparse the history is now
data_all[1]
```

```python
# Which non-LEXIS history citations are not present in our unannotated dataset?
missing_citations = non_lexis_history_citations.difference(all_citations)
with open('missing_citations.txt', 'w') as f:
    for mc in missing_citations:
        f.write(mc + '\n')
```

```python
# Create pairs of casebodies, each row having the parent case and one of its prior history cases
records = {'source_citation': [], 'source_opinion': [], 'history_citation': [], 'history_opinion': []}
for data in data_all:
    citation = data['citation']
    opinion = exploded_df_small[exploded_df_small['citations'] == citation].casebody.values[0]['data']['opinions'][0]['text']
    opinion = ' '.join(opinion.split()[:512])
    for prior in data['priors']:
        records['source_citation'].append(citation)
        records['source_opinion'].append(opinion)
        records['history_citation'].append(prior)
        hist_opinion = exploded_df_small[exploded_df_small['citations'] == prior].casebody.values[0]['data']['opinions'][0]['text']
        hist_opinion = ' '.join(hist_opinion.split()[:512])
        records['history_opinion'].append(hist_opinion)
pair_df = pd.DataFrame.from_records(records, columns=['source_citation', 'source_opinion', 'history_citation', 'history_opinion'])
pair_df.to_csv('case_history_pairs.csv', index=False)
```
